```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error
import kernel as krnl
from deflected_subgradient import solveDeflected
from SVR import SVR
import matplotlib.pyplot as plt
import random
from svr_grid_search import Gridsearch
import time
```

```python
num_inputs = 200
```

```python
# create lin function
lin_x = np.vstack(np.arange(1,num_inputs+1,1))
lin_y = np.array([3*lin_x[i]+random.uniform(-5,5) for i in range(len(lin_x))])
plt.scatter(lin_x,lin_y)
plt.show()
```

```python
# create poly function
poly_x = np.vstack(np.arange(-num_inputs/2,num_inputs/2 + 1,1))
degree = 2
noising_factor = 0.1
poly_y = [xi**degree for xi in poly_x]
poly_y = [ yi + noising_factor * (np.random.rand()*yi) for yi in poly_y]
poly_y=np.array(poly_y, dtype=np.float64)
plt.scatter(poly_x,poly_y)
plt.show()
```

```python
# create sig function
sig_x = np.vstack(np.arange(0,num_inputs,1))
noising_factor = 0.02
gam = 3
sig_y = np.array([np.exp(-xi/2)*(10*np.cos(xi) + 10*np.sin(xi)) for xi in sig_x])
plt.scatter(sig_x,sig_y)
plt.show()
```

```python
# create rbf function
rbf_x = np.vstack(np.arange(-num_inputs/2,num_inputs/2,1))
noising_factor = 0.1
rbf_y = np.array([np.exp(-xi**2/2) for xi in rbf_x])
rbf_y = np.array([ yi + noising_factor * (np.random.rand()*yi) for yi in rbf_y])
plt.scatter(rbf_x,rbf_y)
plt.show()
```

```python
# create random function
rand_x = np.vstack(np.arange(1,num_inputs+1,1))
rand_y = np.array([random.uniform(-num_inputs,num_inputs) for i in range(len(rand_x))])
plt.scatter(rand_x,rand_y)
plt.show()
```

# TRYING TO PLOT TIMINGS

```python
def create_lin_func(num_inputs, num_features=1):
    lin_x = np.vstack(np.arange(1,num_inputs+1,1))
    lin_y = np.array([3*lin_x[i]+random.uniform(-5,5) for i in range(len(lin_x))])
    return lin_x, lin_y

def create_rand_func(num_inputs, num_features):
    rand_x = np.random.rand(num_inputs, num_features)
    rand_y = np.random.rand(num_inputs)
    return rand_x, rand_y
```

```python
input_dims = [1e2, 2e2, 5e2, 1e3, 2e3, 5e3]
timings = []
trial_nums = 5
for i in input_dims:
    timing_sum = 0.0
    lin_x, lin_y = create_lin_func(i)
    for j in range(trial_nums):
        start = time.time()
        svr = SVR("linear", box=0.1)
        svr.fit(lin_x, lin_y, {'eps': 0.01, 'maxiter': 1000.0, 'vareps': 0.1}, optim_verbose=False)
        timing_sum += time.time() - start
    timings.append(timing_sum / trial_nums)
    print(f"Done {i} input_dims in {timings[-1]} sec")
```

```python
plt.plot(input_dims,timings)
plt.scatter(input_dims,timings)
plt.title("LINEAR KERNEL - SCALABILITY ANALYSIS (NUM_INPUTS)")
plt.xlabel("num_inputs")
plt.ylabel("fitting time (s)")
plt.show()
```

```python
feature_dims = [1,2,5,10,20,50,1e2,2e2,5e2,1e3,2e3,5e3]
input_dim = 500
timings_feature = []
rand_x_all, rand_y = create_rand_func(input_dim,int(feature_dims[-1]))
trial_nums = 5
for i in feature_dims:
    timing_sum = 0.0
    rand_x, rand_y = rand_x_all[:,0:int(i)], rand_y
    print(rand_x.shape, rand_x_all.shape)
    for j in range(trial_nums):
        start = time.time()
        svr = SVR("linear", box=0.1)
        svr.fit(rand_x, rand_y, {'eps': 0.01, 'maxiter': 1000.0, 'vareps': 0.1}, optim_verbose=False)
        timing_sum += time.time() - start
    timings_feature.append(timing_sum / trial_nums)
    print(f"Done {i} feature_dims in {timings_feature[-1]} sec")
```

```python
plt.scatter(feature_dims,timings_feature)
plt.plot(feature_dims,timings_feature)
plt.title("LINEAR KERNEL - SCALABILITY ANALYSIS (NUM_FEATURES)")
plt.xlabel("num_features")
plt.ylabel("fitting time")
plt.ylim(10,15)
plt.show()
```
