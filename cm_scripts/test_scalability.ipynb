```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error
import kernel as krnl
from deflected_subgradient import solveDeflected
from SVR import SVR
import matplotlib.pyplot as plt
import random
from svr_grid_search import Gridsearch
import time
```

```python
num_inputs = 5e2
```

```python
# create lin function
lin_x = np.vstack(np.arange(1,num_inputs+1,1))
lin_y = np.array([3*lin_x[i]+random.uniform(-5,5) for i in range(len(lin_x))])
plt.scatter(lin_x,lin_y)
plt.show()
```

```python
# create poly function
poly_x = np.vstack(np.arange(-num_inputs/2,num_inputs/2 + 1,1))
degree = 2
noising_factor = 0.1
poly_y = [xi**degree for xi in poly_x]
poly_y = [ yi + noising_factor * (np.random.rand()*yi) for yi in poly_y]
poly_y=np.array(poly_y, dtype=np.float64)
plt.scatter(poly_x,poly_y)
plt.show()
```

```python
# create sig function
sig_x = np.vstack(np.arange(0,num_inputs,1))
noising_factor = 0.02
gam = 3
sig_y = np.array([np.exp(-xi/2)*(10*np.cos(xi) + 10*np.sin(xi)) for xi in sig_x])
plt.scatter(sig_x,sig_y)
plt.show()
```

```python
# create rbf function
rbf_x = np.vstack(np.arange(-num_inputs/2,num_inputs/2,1))
noising_factor = 0.1
rbf_y = np.array([np.exp(-xi**2/2) for xi in rbf_x])
rbf_y = np.array([ yi + noising_factor * (np.random.rand()*yi) for yi in rbf_y])
plt.scatter(rbf_x,rbf_y)
plt.show()
```

```python
# create random function
rand_x = np.vstack(np.arange(1,num_inputs+1,1))
rand_y = np.array([random.uniform(-num_inputs,num_inputs) for i in range(len(rand_x))])
plt.scatter(rand_x,rand_y)
plt.show()
```
